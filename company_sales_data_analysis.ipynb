```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
!pip install seaborn
import seaborn as sns
```

```python
# Read the data from the provided link

data = pd.read_csv("D:\\NPower\\Projects\\Company_sales\\company_sales_data.csv")
data.head()
```

```python
# Exercise 1: Read Total profit of all months and show it using a line plot
```

```python
# Total profit data provided for each month. Generated line plot must include the following properties: â€“
# X label name = Month Number
# Y label name = Total profit

ax = data.plot(x='month_number', y='total_profit')
ax.set_xlabel("Month Number")
ax.set_ylabel("Total profit")
plt.show()
```

```python
# Exercise 2: Get total profit of all months and show line plot with the following Style properties
# Generated line plot must include following Style properties: 
```

```python
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3

data.plot(x = "month_number", y = "total_units", markerfacecolor='red', linestyle='dashed', marker='o',
     linewidth=3)
plt.show()
```

```python
# Exercise 3: Read all product sales data and show it using a multiline plot
# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

monthly_units = data.groupby('month_number')[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()

# Generate the multiline plot
monthly_units.plot()

# Add labels and title to the plot
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Number of Units Sold per Month for Each Product')

# Display the plot
plt.show()
```

```python
sns.lineplot(data[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']])
```

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
# Also, add a grid in the plot. gridline style should â€œâ€“â€œ.

# Generate the scatter plot
plt.scatter(data['month_number'], data['toothpaste'])

# Add gridlines to the plot
plt.grid(linestyle='--')

# Add labels and title to the plot
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales per Month')

# Display the plot
plt.show()
```

```python
sns.scatterplot(data, x = "month_number", y = "toothpaste")
```
